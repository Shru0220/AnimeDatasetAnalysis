```python
#Replacing values in numeric columns

df['Rank'].replace('UNKNOWN', 0, inplace=True)
df['Score'].replace('UNKNOWN', 0, inplace=True)
```

```python
df['Aired'] = pd.to_datetime(df['Aired'], errors='coerce')
df['Premiered'] = pd.to_datetime(df['Premiered'], errors='coerce')

# Convert numeric columns
df['Score'] = pd.to_numeric(df['Score'], errors='coerce')
df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce')
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce')
df['Popularity'] = pd.to_numeric(df['Popularity'], errors='coerce')
df['Favorites'] = pd.to_numeric(df['Favorites'], errors='coerce')
df['Scored By'] = pd.to_numeric(df['Scored By'], errors='coerce')
df['Members'] = pd.to_numeric(df['Members'], errors='coerce')
```

```python
# Drop duplicates based on 'anime_id' or 'Name'
df.drop_duplicates(subset=['anime_id'], inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Replace blank entries with NaN
df[['Episodes', 'Aired', 'Premiered', 'Scored By']] = df[['Episodes', 'Aired', 'Premiered', 'Scored By']].replace('', pd.NA)
```

```python
# Fill blanks in 'Episodes' with 0 (numeric column)
df['Episodes'].fillna(0, inplace=True)

# Fill blanks in 'Aired' and 'Premiered' with a placeholder text, such as "Unknown"
df['Aired'].fillna(0, inplace=True)
df['Premiered'].fillna(0, inplace=True)

# Fill blanks in 'Scored By' with 0 (numeric column)
df['Scored By'].fillna(0, inplace=True)
```
